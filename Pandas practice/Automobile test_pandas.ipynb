## question 5- Based on new regulations, companies decided to change the prices of the car. The new update price is calculated as - if the engine is in front, the price will be the same else if the engine is in the rear, the price will be doubled. Add a new column with the updated prices.

```python
df['engine-location'].unique()
```

```python
def value_data(price,engine):
    if engine=='front':
        return price
    elif engine=='rear':
        return price*2
```

```python
df['price_update']=df.apply(lambda x: value_data(x['price'],x['engine-location']),axis=1)
```

## question 7- Create a new column which stores the number of doors in a car as integers.

```python
df['num-of-doors'].unique()
```

```python
def value(doors):
    if doors =='two':
        return 2
    elif doors=='four':
        return 4
    else:
        return np.nan
```

```python
df['number_of_doors_numeric']= df['num-of-doors'].apply(value)
```
